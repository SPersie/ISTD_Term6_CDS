```python
import pandas as pd
dataframe = pd.read_csv('scottish_hills.csv')
print(dataframe.head(10))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x = dataframe.Height
y = dataframe.Latitude
z = dataframe.Longitude
```

```python
plt.scatter(x, y)
plt.show()
```

```python
from scipy.stats import linregress
stats = linregress(x, y)

m = stats.slope
b = stats.intercept
```

```python
plt.scatter(x, y)
plt.plot(x, m * x + b, color="red")
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(x, y, marker='x')

plt.plot(x, m * x + b, color="red", linewidth=3)

plt.xlabel("Height (m)", fontsize=20)
plt.ylabel("Latitude", fontsize=20)

plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
```

```python
import numpy as np
colors = np.random.rand(len(y))
plt.scatter(y, z, s = (x-900), c=colors, alpha=0.5)
```

```python
plt.hist(x, bins=25, normed=True)
plt.show()
```

```python
import numpy as np
plt.style.use('seaborn-pastel')

shifted_x = x -100
fig, ax = plt.subplots()
ax.hist(x, bins=25, normed=True, histtype="stepfilled", alpha=0.8, label='Height')
ax.hist(shifted_x, bins=25, normed=True, histtype="stepfilled", alpha=0.8, label='Height - 100')

ax.set_ylabel('Normalised distribution')
ax.set_xlabel("Height")
ax.legend()
```

```python
labels = ['EU', 'US', 'AS']
meateaters = [122, 135, 80]
vegetarians = [40, 43, 23]

index = np.arange(len(labels))

fig, ax = plt.subplots()

bar_width = 0.35
opacity = 0.6

rects1 = ax.bar(index, meateaters, bar_width,
               alpha=opacity, color='b',
               label='Meat eaters')

rects2 = ax.bar(index + bar_width, vegetarians, bar_width,
               alpha=opacity, color='r',
               label='Vegetarians')

ax.set_xlabel('Continent')
ax.set_ylabel('CO2 emissions')
ax.set_title('CO2 emissions per continent, oer diet')

ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(labels)
ax.legend()
```

```python
def fahreheit2celsius(temp):
    return (5. / 9.) * (temp - 32)

temperature = [100, 102, 106, 105, 90, 85, 89, 100, 102, 103, 108, 100, 102, 106, 105, 90, 85, 85, 89, 100, 102, 103, 108]
```

```python
fig, ax_f = plt.subplots()
ax_c = ax_f.twinx()

ax_f.plot(temperature)
ax_f.set_xlim(0, 24)

y1, y2 = ax_f.get_ylim()
ax_c.set_ylim(fahreheit2celsius(y1), fahreheit2celsius(y2))
ax_c.figure.canvas.draw()

ax_f.set_title('Two scales: Fahrenheit and Celsius')
ax_f.set_ylabel('Fahrenheit')
ax_c.set_ylabel('Celsius')
ax_f.set_xlabel('Time (hour of day)')
```
