```python
import pandas as pd
dataframe = pd.read_csv("school_earnings.csv")
print(dataframe.head(10))
%matplotlib inline
import matplotlib.pyplot as plt
x = dataframe.Men
y = dataframe.Women
z = dataframe.Gap
```

```python
# Histograms for women
```

```python
plt.hist(y, bins=25, normed=True)
plt.show()
```

```python
# Histograms for both men and women
# Dark background and label the axes
# Label the colors of the histogram (Women and Men)
```

```python
import numpy as np
import seaborn as sns
plt.style.use("seaborn-pastel")
fig, ax = plt.subplots()
ax.set_facecolor('black')
ax.hist(x, bins=25, normed=True, histtype="stepfilled", alpha=0.8, label='Men')
ax.hist(y, bins=25, normed=True, histtype="stepfilled", alpha=0.8, label='Women')
ax.legend(prop={'size': 10})

ax.set_ylabel('Normalised distribution')
ax.set_xlabel('Earning')
```

```python
# Create a bar chart that lists the salary for women (y-axis) for each school
```

```python
labels = dataframe.School
index = np.arange(len(labels))

fig, ax = plt.subplots()
bar_width = 0.5
opacity = 0.6

rects2 = ax.bar(index, y, bar_width,
               alpha=opacity, color='r',
               label='Women')

ax.set_xlabel('Schools')
ax.set_ylabel('Salary')

ax.set_xticks(index)
ax.set_xticklabels(labels)
ax.legend()
```

```python
# bar charts for both women and men
```

```python
fig, ax = plt.subplots()
bar_width = 0.5
opacity = 0.6

rects1 = ax.bar(index, x, bar_width,
               alpha=opacity, color='b',
               label='Men')

rects2 = ax.bar(index + bar_width, y, bar_width,
               alpha=opacity, color='r',
               label='Women')

ax.set_xlabel('Schools')
ax.set_ylabel('Salary')

ax.set_xticks(index + bar_width/2)
ax.set_xticklabels(labels, rotation= 90)
ax.legend()
```
